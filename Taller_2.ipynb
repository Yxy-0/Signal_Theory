# **Ejercicio 1: Sistema de Biblioteca**
Crea una clase Libro que contenga atributos como título, autor, año de publicación y estado
(disponible o prestado). Luego, desarrolla una clase Biblioteca que gestione una colección de
objetos Libro. Esta clase debe incluir métodos para:

Añadir un nuevo libro al catálogo.
Prestar un libro a un usuario (cambiar el estado a prestado).
Devolver un libro (cambiar el estado a disponible).
Buscar libros por autor o título.
Mostrar los libros disponibles para préstamo.




```python
class Libro:  # Constructor de la clase Libro
    def __init__(self,titulo,autor,publicacion,estado="disponible"):
      self.titulo=titulo # Título del libro
      self.autor=autor  # autor del libro
      self.publicacion=publicacion #año de publicacion
      self.estado="disponible"

    def __str__(self):
      return f"-Título: {self.titulo}, Autor: {self.autor}, Año de publicación: {self.publicacion}, Estado: {self.estado}" # Método especial para representar el objeto como una cadena

class Biblioteca:  # Constructor de la clase Biblioteca
    def __init__(self):
      self.libreria= []  # Lista para almacenar los libros

    def agregar_libro(self, libro): # Método para agregar un libro a la biblioteca
      self.libreria.append(libro)

    def prestar_libro(self, titulo): # Método para prestar un libro
      for libro in self.libreria:
        if libro.titulo==titulo:
          if libro.estado=="disponible": #Si el libro está disponible, cambia el estado a prestado, sino dice que ya se encuentra prestado o si no existe en la biblioteca
            libro.estado="prestado"
            print ("**el libro", titulo, " fue prestado**")
          else:
            print ("**el libro", titulo,"  ya se encuentra prestado**")
          return
      print ("**el libro",titulo, "no existe en la biblioteca**")

    def devolver_libro(self, titulo):  # Método para devolver un libro prestad
      for libro in self.libreria:
        if libro.titulo==titulo:
          if libro.estado=="prestado": # Cambiar el estado del libro a disponible si se encuentra prestado
            libro.estado="disponible"
            print ("**el libro" ,titulo, "fue devuelto exitosamente**")
          else:
            print ("**el libro" ,titulo, "está disponible**")
          return
      print ("**el libro" ,titulo, "no existe en la biblioteca**")


    def buscar_autor(self, autor): # Método para buscar libros por autor
      encontrados = False
      print("**Se encontraron los siguientes libros del  autor: ", autor)
      for libro in self.libreria:
          if libro.autor == autor:
            print(libro)
            encontrados = True
      if not encontrados:
        print("**No se encontraron libros del autor",autor," en la biblioteca.")

    def buscar_titulo(self, titulo): # Método para buscar libros por título
        encontrados = False
        print("**Se encontraron los siguientes libros con el titulo: ", titulo)
        for libro in self.libreria:
            if libro.titulo == titulo:
                print(libro)
                encontrados = True
        if not encontrados:
          print("**No se encontraron libros con el titulo",titulo," en la biblioteca.")


    def mostrar_disponibles(self): # Método para mostrar libros disponibles
      disponibles = False
      print("Libros disponibles:")
      for libro in self.libreria:
          if libro.estado == "disponible":
              print(libro)
              disponibles = True
      if not disponibles:
          print("No hay libros disponibles para préstamo en la biblioteca.")


# Crear una biblioteca
biblioteca = Biblioteca()

libro1 = Libro("Orgullo y prejuicio", "Jane Austen", 1813, "disponible")
libro2 = Libro("Los juegos del hambre", "Suzanne Collins", 2008, "disponible")
libro3 = Libro("Harry Potter y la piedra filosofal", "J.K. Rowling", 1997, "disponible")
libro4 = Libro("Harry Potter y la cámara secreta", "J.K. Rowling", 1998, "disponible")
libro5 = Libro("Harry Potter y el prisionero de Azkaban", "J.K. Rowling", 1999, "disponible")

# Agregar libros a la biblioteca
biblioteca.agregar_libro(libro1)
biblioteca.agregar_libro(libro2)
biblioteca.agregar_libro(libro3)
biblioteca.agregar_libro(libro4)
biblioteca.agregar_libro(libro5)


biblioteca.mostrar_disponibles() #muestra los libros disponibles

biblioteca.prestar_libro("Orgullo y prejuicio") #presta el libro

biblioteca.devolver_libro("Orgullo y prejuicio") #devuelve el libro

biblioteca.prestar_libro("Harry Potter y la cámara secreta") #presta el libro

biblioteca.buscar_autor("J.K. Rowling") #busca el libro por autor

biblioteca.buscar_titulo("Los juegos del hambre") #busca el libro por el titulo

```

# **Ejercicio 2: Sistema de Reservas de Hotel**
Define una clase Habitacion con atributos como número de habitación, tipo (simple, doble, suite),
precio por noche y estado (ocupada o disponible). Luego, crea una clase Hotel que administre
varias habitaciones. La clase Hotel debe ofrecer funcionalidades para:

Registrar una nueva habitación.
Reservar una habitación, cambiando su estado a ocupado.
Liberar una habitación después de la estancia, cambiando su estado a disponible.
Calcular el costo de la estancia de un cliente.
Mostrar todas las habitaciones disponibles.

```python
class Habitacion:
  def __init__(self,numero,tipo,precio,estado="disponible"): # Constructor de la clase Habitacion
    self.numero=numero # Número de la habitación
    self.tipo=tipo # Tipo de habitación
    self.precio=precio # Precio por noche
    self.estado=estado # Estado de la habitación
  def __str__(self):
    return f"-Habitacion numero: {self.numero}, tipo {self.tipo}, precio por noche: {self.precio}, Estado: {self.estado}"  # Método especial para representar el objeto como una cadena

class Hotel:
  def __init__(self): # Constructor de la clase Hotel
    self.habitaciones= [] # Lista para almacenar las habitaciones

  def registrar_habitacion(self, numero): # Método para registrar una habitación en el hotel
    self.habitaciones.append(numero)

  def reservar_habitacion(self, numero):  # Método para reservar una habitación
    for habitacion in self.habitaciones:
      if habitacion.numero == numero:
        if habitacion.estado == "disponible": # Cambiar el estado de la habitación a ocupada si se encuentra disponible, sino dice que la habitacion esta ocupada o que no se encuentra
          habitacion.estado = "ocupada"
          print("**Habitación",numero,"se ha reservado exitosamente**")
        else:
          print("**La habitación",numero," se encuentra ocupada**")
        return
    print("**No se encontró la habitación",numero, "en el hotel.**")

  def devolver_habitacion(self, numero): # Método para devolver una habitación
    for habitacion in self.habitaciones:
      if habitacion.numero == numero:
        if habitacion.estado == "ocupada": # Cambiar el estado de la habitación a disponible si se encontraba ocupada
          habitacion.estado = "disponible"
          print("**Habitación",numero,"se ha devuelto exitosamente**")
        else:
          print("**La habitación",numero,"se encuentra disponible**")
        return
    print("**No se encontró la habitación",numero, "en el hotel.**")

  def costo_noche(self,numero,noches):  # Método para calcular el costo total de una estadía
    for habitacion in self.habitaciones:
      if habitacion.numero == numero and habitacion.estado =="ocupada": #si encuentra el numero de la habitacion y se encuentra ocupada por el residente, se realiza el costo
        costo= habitacion.precio * noches
        print("**El costo total por:", noches, " noches, de la habitacion:", numero , "es de: ",costo, "**")

  def mostrar_disponibles(self): # Método para mostrar las habitaciones disponibles
      disponibles = False
      print("**Habitaciones disponibles:")
      for habitacion in self.habitaciones:
        if habitacion.estado== "disponible":
          print(habitacion)
          disponibles = True
      if not disponibles:
          print("No hay libros disponibles para préstamo en la biblioteca.")


#creando el hotel

hotel=Hotel()


habitacion101=Habitacion(101, "simple", 20000, "disponible")
habitacion102=Habitacion(102, "doble", 60000, "disponible")
habitacion103=Habitacion(103, "suite", 120000, "disponible")

# Registrando las habitaciones en el hotel
hotel.registrar_habitacion(habitacion101)
hotel.registrar_habitacion(habitacion102)
hotel.registrar_habitacion(habitacion103)

hotel.mostrar_disponibles()  # Muestra las habitaciones disponible


hotel.reservar_habitacion(101) # Reserva la habitacion
hotel.reservar_habitacion(102) # Reserva la habitacion
hotel.costo_noche(102,7) # Da el costo de la habitacion por las noches que se desea quedar

hotel.devolver_habitacion(101) # devuelve la habitacion


```

# **Ejercicio 3: Sistema de Gestión de Empleados**
Implementa una clase Empleado con atributos como nombre, identificación, puesto y salario.
Posteriormente, desarrolla una clase Departamento que contenga una lista de empleados. Esta
clase debe ser capaz de:

Añadir un nuevo empleado.
Eliminar un empleado por su identificación.
Aumentar el salario de un empleado por un porcentaje dado.
Mostrar la información de todos los empleados.
Calcular el gasto total en salarios dentro del departamento.

```python
class Empleado:
  def __init__(self,nombre,identificacion,puesto,salario):  # Constructor de la clase Empleado
    self.nombre=nombre  # Nombre del empleado
    self.identificacion=identificacion # Identificación del empleado
    self.puesto=puesto # Puesto del empleado
    self.salario=salario # Salario del empleado

  def __str__(self):
    return f"Nombre: {self.nombre}, Identificación: {self.identificacion}, Puesto: {self.puesto}, Salario: ${self.salario}"  # Método especial para representar el objeto como una cadena

class Departamento:
  def __init__(self): # Constructor de la clase Departamento
    self.empleados= [] # Lista para almacenar los empleados del departamento

  def nuevo_empleado(self, empleado): # Método para añadir un nuevo empleado al departamento
    self.empleados.append(empleado)

  def eliminar_empleado(self, identificacion): # Método para eliminar un empleado por su identificación
    for empleado in self.empleados:
      if empleado.identificacion == identificacion: # si encuentra la identificacion del empleado, se elimina de los datos del departamento, sino dice que no se encontro la identificacion
          self.empleados.remove(empleado)
          print("**El empleado con identificacion",identificacion,"se ha eliminado exitosamente**")
          return
    print("**No se encontro el empleado con identificacion: ",identificacion, "en la base de datos**")

  def aumentar_salario(self, identificacion, aumento):  # Método para aumentar el salario de un empleado por un porcentaje dado
    for empleado in self.empleados:
      if empleado.identificacion == identificacion:
        empleado.salario += empleado.salario*(aumento/100)
        print("**El empleado con identificacion",identificacion,"se le ha realizado un aumento del", aumento, ", su nuevo salario es:", empleado.salario)
        return
    print("**No se encontro el empleado con identificacion: ",identificacion, "en la base de datos**")

  def gasto_total(self): # Método para calcular el gasto total en salarios dentro del departamento
    gasto= sum(empleado.salario for empleado in self.empleados)
    print("El gasto total en salarios dentro del departamento es de:", gasto)


  def informacion_empleados(self):  # Método para mostrar la información de todos los empleados
    print("**La informacion de empleados actual es:")
    for empleado in self.empleados:
      print(empleado)


#creando el departamento

departamento = Departamento()

empleado1 = Empleado("Yuliana", 102301, "Presidente", 1850000)
empleado2 = Empleado("Isabella", 342453, "Gerente", 60000)
empleado3 = Empleado("Daniel", 239086, "Analista", 10000)

departamento.nuevo_empleado(empleado1) # Añade empleado1 al departamento
departamento.nuevo_empleado(empleado2) # Añade empleado2 al departamento
departamento.nuevo_empleado(empleado3) # Añade empleado3 al departamento


departamento.informacion_empleados() # Muestra la información actual de los empleados

departamento.eliminar_empleado(342453) # Eliminando un empleado por su identificación

departamento.informacion_empleados() # Muestra la información actual de los empleados

departamento.aumentar_salario(239086, 50) # Aumenta el salario de un empleado por un porcentaje

departamento.informacion_empleados() # Muestra la información actual de los empleados

departamento.gasto_total() # Calculando el gasto total en salarios dentro del departamento
```

# **Ejercicio 4: Tienda en Línea**

Crea una clase Producto con atributos tales como identificador, nombre, precio y cantidad en
stock. Después, diseña una clase CarritoDeCompras que permita a los usuarios agregar o eliminar
productos, y calcular el total de su compra. La clase CarritoDeCompras debe ofrecer métodos
para:

Añadir un producto al carrito.
Eliminar un producto del carrito.
Mostrar los productos en el carrito.
Calcular el total de la compra.
Aplicar descuentos a productos específicos o al total de la compra.

```python
class producto:
  def __init__(self,identificador,nombre,precio,stock):  # Constructor de la clase Producto
    self.identificador=identificador # Identificador del producto
    self.nombre=nombre # Nombre del producto
    self.precio=precio # Precio del producto
    self.stock=stock # Cantidad en stock del producto

  def __str__(self):
    return f"-ID: {self.identificador}, Nombre: {self.nombre}, Precio: ${self.precio}, Cantidad en stock: {self.stock}" # Método especial para representar el objeto como una cadena

class carrito_compras:
  def __init__(self): # Constructor de la clase CarritoCompras
    self.productos = [] # Lista para almacenar los productos en el carrito

  def agregar_producto(self, producto): # Método para agregar un producto al carrito
    self.productos.append(producto)
    print("-se añadio al carrito:", producto.nombre)

  def eliminar_producto(self, identificador): # Método para eliminar un producto del carrito por su identificador
    for producto in self.productos:
      if producto.identificador == identificador: #si el identificador se encuentra, se elimina del carrito, sino se avisa que no se encontró ninguno con ese ifentificador
        self.productos.remove(producto)
        print ("**Producto con identificador ", identificador, "se ha eliminado del carrito.**")
        return
    print ("No se encontró ningún producto con identificador", identificador, " en el carrito.")

  def total_compra(self):  # Método para calcular el total de la compra
    total = sum(producto.precio for producto in self.productos)
    print("**Total de la compra: $",total)
    return

  def descuento_producto(self, identificador, descuento):  # Método para aplicar un descuento a un producto por su identificador
    for producto in self.productos:
      if producto.identificador == identificador : #si se encuentra el producto por su identificador se realiza el descuento, sino avisa que no encontró el producto
        producto.precio -= producto.precio * (descuento / 100)
        print("**se ha aplicado un descuento de:",descuento,"% al producto con identificador: ", identificador)
        return
    print("No se encontro el producto con el identificador: ", identificador)

  def descuento_total(self, descuento): # Método para aplicar un descuento total al carrito
    for producto in self.productos:
        producto.precio -= producto.precio * (descuento / 100)
    print("**se ha aplicado un descuento de:",descuento,"% a todo el carrito de compra**")

  def mostrar_productos(self):  # Método para mostrar los productos en el carrito
    print("**Los productos en el carrito de compra son:")
    for producto in self.productos:
      print(producto)

# crear carrito
carrito = carrito_compras()

producto1 = producto(23451, "Camiseta", 20000, 100)
producto2 = producto(23218, "Pantalón", 150000, 50)
producto3 = producto(97803, "Zapatos", 50000, 30)
producto4 = producto(10393, "medias", 5000, 20)

carrito.agregar_producto(producto1) #agrega producto1 al carrito
carrito.agregar_producto(producto2) #agrega producto2 al carrito
carrito.agregar_producto(producto3) #agrega producto3 al carrito
carrito.agregar_producto(producto4) #agrega producto4 al carrito


carrito.mostrar_productos() # Muestra los productos que hay en el carrito

carrito.eliminar_producto(23218) #elimina el producto 23451 del carrito

carrito.mostrar_productos() #muestra lo que esta en el carrito

carrito.descuento_producto(23451, 10) # realiza un descuento de 10% al producto 23451
carrito.descuento_producto(97803, 60) # realiza un descuento del 60% al producto 97803

carrito.descuento_total(5) #realiza un descuento del 5% a todo el carrito

carrito.total_compra() #da el total de la compra
```
